```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing Data
stroke_data = pd.read_csv("./input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
```

```python
#Dropping NA
stroke_data = stroke_data.dropna()
```

```python
#Splitting X and y
X = stroke_data.drop(["id","stroke"],axis=1)
```

```python
y = stroke_data["stroke"]
```

```python
X.head()
```

```python
# from sklearn.preprocessing import OneHotEncoder
# from sklearn.compose import ColumnTransformer
# categorical_features = ["gender","age","hypertension","heart_disease","ever_married","avg_glucose_level","bmi","smoking_status"]
# one_hot = OneHotEncoder()
# transformer = ColumnTransformer([("one_hot",
#                                 one_hot,
#                                 categorical_features)],remainder = "passthrough")
# transformed_X = transformer.fit_transform(X)
# transformed_X


#Using dummies to convert to numerical
dummies = pd.get_dummies(stroke_data[["gender","age","hypertension","heart_disease","ever_married","avg_glucose_level","bmi","smoking_status"]])
dummies
```

```python
#using randomforestclassifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.get_params()
```

```python
#Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dummies,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
#RFC score
clf.score(X_test,y_test)
```

```python
clf.score(X_train,y_train)
```

```python
#Trying Linear SVC as per chart and it's score
from sklearn.svm import LinearSVC
np.random.seed(42)

X_train,X_test,y_train,y_test = train_test_split(dummies,y,test_size=0.2)

svc=LinearSVC()
svc.fit(X_train,y_train)
svc.score(X_test,y_test)
```
