```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
salesdf = pd.read_csv("/content/drive/MyDrive/data/sql_runner_xmmq3tzwsb9dyg_2022-06-25_04-59-03.csv")
```

```python
salesdf.head(3)
```

```python
salesdf['calmonth'] = pd.to_datetime(salesdf['calmonth'], format='%Y%m')
```

```python
salesdf.drop(columns=["store_type","store_code","address","article_number","prod_division","prod_group","marketing_division","key_category_cluster","key_category","zrt_year","article_flag"],inplace=True)
```

```python
salesdf = salesdf[salesdf['prod_group_name'].notna()]
salesdf[['net_qty', 'net_sales','stock_qty','stock_val','net_qty_py','stock_qty_py','stock_val_py','net_sales_py']] = salesdf[['net_qty', 'net_sales','stock_qty','stock_val','net_qty_py','stock_qty_py','stock_val_py','net_sales_py']].fillna(value=0)
salesdf[['article_name','gender']] = salesdf[['article_name','gender']].fillna(value='NA')
```

```python
grouped = salesdf[['calmonth','store_type_name','currency_code','net_sales']].groupby(['calmonth','store_type_name','currency_code']).sum()
grouped
```
